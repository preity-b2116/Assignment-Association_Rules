```python
import pandas as pd
import seaborn as sns
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
book = pd.read_csv('D:/ExcelR/Data Science/Assignments related/CSV/book.csv')
book.head()
```

```python
book.shape
```

```python
book.info()
```

## Apriori Algorithm

### for Minimum Support = 0.1

```python
frequent_itemsets1 = apriori(book, min_support=0.1, use_colnames=True)
frequent_itemsets1
```

```python
rules1 = association_rules(frequent_itemsets1, metric="confidence", min_threshold=0.7)
rules1
```

#####  An leverage value of 0 indicates independence. Range will be [-1 1]
A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules1.sort_values('confidence',ascending = False)[0:20]
```

```python
rules1[rules1.confidence>0.8]
```

## Apriori Algorithm

### for Minimum Support = 0.9

```python
frequent_itemsets2 = apriori(book, min_support=0.09, use_colnames=True)
frequent_itemsets2
```
